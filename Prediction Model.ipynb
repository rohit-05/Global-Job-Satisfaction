```python
def getAccuracy():
    diff = []
    diff = y_test - prediction_svm
    sumOfDiff = 0
    for each in diff:
        sumOfDiff += abs(each)
        accuracy = 1-((sumOfDiff/len(Dataset))/5) #gets average different among test and prediction and divides by 5. Then subtracts that from 1 to find the percent accuracy   
    return accuracy
```

### SGDClassifier

```python
resultsCorrect = []
resultsAcc = []

for i in range(0,50):
    pipeline = Pipeline([('sgd', SGDClassifier(loss='hinge',
                                               alpha=1e-2,
                                               penalty='l2',
                                               max_iter=5,
                                               random_state=None))])

    pipeline = pipeline.fit(X_train, y_train)
    prediction_svm = pipeline.predict(X_test)
    predictions = pipeline.predict(X_test).tolist()
    resultsCorrect.append(np.mean(prediction_svm == y_test))
    resultsAcc.append(getAccuracy())
    
print(sum(resultsCorrect)/len(resultsCorrect)) #0.41161384869
print(sum(resultsAcc)/len(resultsAcc)) #0.968030774867
```
